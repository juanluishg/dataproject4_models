```python
from os import listdir, getcwd
from os.path import isfile, join
```

```python
detector = MTCNN()
```

```python
def get_all_faces(img):
    results = detector.detect_faces(img)
    caras = []
    
    for i in results:
        # caja de la cara
        x, y, w, h = i['box']

        cara = img[y: y+h, x: x+w]
        
        caras.append(cara)
        
    return caras, results
```

```python
def read_image(path):
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
def get_face_from_path(path):
    image = read_image(path)
    return get_all_faces(image)
```

```python
bbox_df = pd.DataFrame()
```

```python
folders = ['ade20k', 'emodb_small', 'framesdb', 'mscoco']
```

```python
%time
for f in folders:
    print('Reading folder: ',f)
    abs_path = getcwd()
    abs_path = abs_path.replace('\\', '/')
    path = '/raw_data/EMOTIC_DATASET/emotic/' + f + '/images/'
    all_filenames = listdir(abs_path + path)
    for n in all_filenames:
        bboxs = get_face_from_path(abs_path + path + n)
        bbox_df = bbox_df.append(pd.DataFrame({'filename':n, 'bboxs': bboxs}))
```
