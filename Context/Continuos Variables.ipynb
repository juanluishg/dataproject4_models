```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from fastai.vision import *
import torch
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../datasets/emotics_annotations_clean_v4.csv")
df.head()
```

```python
df.columns
```

```python
df = df[["filename", "folder", "valence", "arousal", "dominance"]]
y_columns = ["valence","arousal", "dominance"]
```

```python
std = StandardScaler()
y = std.fit_transform(df[y_columns])
y = pd.DataFrame(y ,columns=y_columns)
```

```python
X = df.drop(y_columns, axis=1)
```

```python
X["name"] = X["folder"] + "/" + X["filename"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=30)
```

```python
df_fastai = pd.DataFrame()
df_fastai["name"] = X_train["name"]
df_fastai[y_columns] = y_train
```

```python
df_fastai.head()
```

```python
df_final = pd.DataFrame()
path = '../raw_data/EMOTIC_DATASET/emotic/'
for index, row in df_fastai.iterrows():
    if os.path.isfile(path+row['name']):
        df_final = df_final.append(row)
    else:
        print('No hay na')
```

```python
cols = ["name", "valence","arousal", "dominance"]
df_final = df_final[cols]
df_final.head()
```

```python
df_final = df_final.dropna()
```

```python
#path = Path('../raw_data/EMOTIC_DATASET/emotic/')
path = '../raw_data/EMOTIC_DATASET/emotic/'
tfms = get_transforms(do_flip=False)
data_utk = ImageList.from_df(df_final, path, cols=['name'], folder ='.').split_by_rand_pct(0.2, seed=42).label_from_df(label_cls=FloatList).transform(tfms, resize_method=ResizeMethod.CROP, padding_mode='border', size=256).databunch(bs=64*2,num_workers=0, pin_memory=False).normalize(imagenet_stats)
```

```python
data_utk.show_batch(4, figsize=(12,12))
```
