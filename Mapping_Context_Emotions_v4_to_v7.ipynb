```python
df = pd.read_csv("emotics_annotations_clean_v4.csv")
df.head()
```

```python
mapping = {"Disapproval": "Angry", "Annoyance": "Angry", "Anger": "Angry", "Disquietment": "Angry",
"Disconnection": "Disgust", "Fatigue": "Disgust", "Embarrassment": "Disgust", "Aversion": "Disgust",
"Doubt/Confusion": "Fear", "Fear" : "Fear",
"Confidence": "Happy", "Happiness": "Happy", "Pleasure": "Happy", "Excitement": "Happy", "Yearning" : "Happy",
"Sadness": "Sad", "Pain": "Sad", "Suffering" : "Sad", 
"Anticipation": "Surprise", "Surprise" : "Surprise",
"Peace": "Neutral", "Affection": "Neutral", "Esteem": "Neutral", "Engagement": "Neutral", "Sympathy": "Neutral", "Sensitivity" : "Neutral"
}
```

```python
df["cats"] = df.categories.str.split(",")
```

```python
for i, row in df.iterrows():
  phoenetic = [mapping[cat] for cat in row["cats"]]
  df.loc[i, "mapping_cats"] = ','.join(np.unique(np.array(phoenetic)))
```

```python
df.head()
```

```python
df.to_csv("emotics_annotations_clean_v6.csv")
```
